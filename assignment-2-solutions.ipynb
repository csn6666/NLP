```python
import re
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
def extract(resp, splitter, rule):
    r_split = resp.split(splitter)
    results = []
    for r_child in r_split:
        a = re.findall(rule, r_child)
        if a:
            for line in a:
                results.append(list(line))
    return results
```
