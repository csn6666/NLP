```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
def extract_urls(resp, splitter, rule):
    r_split = resp.split(splitter)
    urls = []
    for r_child in r_split:
        a = re.findall(rule, r_child)
        if a:
            for line in a:
                urls.append(list(line))
    for url in urls:
        url[0] = url[0].split('\"')[0]
    return urls
```

```python
headers = {"User-Agent" : 
           "User-Agent:Mozilla/5.0 (compatible; MSIE 9.0; Windows NT 6.1; Trident/5.0;"}
idx = 'https://baike.baidu.com'
```

```python
lines_path = idx + '/item/北京地铁/408485'
r = requests.get(lines_path, 
                 allow_redirects=False,headers=headers,verify=False)
r.encoding='utf-8'
lines = extract_urls(r.text, r"</td>|</a>", r"href=[\"\']/item(.+)[\"\']>(北京地铁.+线)") 
```

```python
stations = {}
for line in lines:
    print(line[1]+'\n')
    line_path = idx + '/item'+line[0]
    r = requests.get(line_path, 
                     allow_redirects=False,headers=headers,verify=False)
    r.encoding='utf-8'
    stations[line[1]] = extract_urls(r.text, r"</td>|</a>", r"<a target=_blank href=[\"\']/item(.+)[\"\']>(\w+站)")
```
